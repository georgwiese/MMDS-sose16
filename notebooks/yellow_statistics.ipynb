# Basic Analysis
This Notebook will compute (and plot) some basic statistics on a sample of the data.

```python
# Initialization

import pandas as pd

%matplotlib inline
pd.set_option('display.mpl_style', 'default') 
pd.set_option('display.line_width', 5000) 
pd.set_option('display.max_columns', 60)

DATA_PATH = "../data/yellow_sample_001.csv"
```

## Data Overview

```python
data = pd.read_csv(DATA_PATH,
                   parse_dates=["Trip_Pickup_DateTime", "Trip_Dropoff_DateTime"],
                   index_col="Trip_Pickup_DateTime")
data["Weekday"] = data.index.weekday
data["Year"] = data.index.year
data["Month"] = data.index.month
data["Hour"] = data.index.hour
data = data.sort()
data
```

## Distribution of Pickup Time

```python
data.Year.value_counts(sort=False).plot(kind="bar", figsize=(15,5), title="Year")
```

```python
data.Month.value_counts(sort=False).plot(kind="bar", figsize=(15,5), title="Month")
```

```python
weekday_counts = data.Weekday.value_counts(sort=False)
weekday_counts.index = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
weekday_counts.plot(kind="bar", figsize=(15,5), title="Weekday")
```

```python
data.Hour.value_counts(sort=False).plot(kind="bar", figsize=(15,5), title="Hour")
```
